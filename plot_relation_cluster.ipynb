```python
import matplotlib.pyplot as plt
import seaborn as sns; sns.set()
import pandas as pd
import os
import numpy as np
```

```python
exp_dir = os.path.join('clean_dataset', 'Current health expenditure (% of GDP).csv')
cluster_dir = os.path.join('cluster_result', 'clustering_result.xlsx')
```

```python
exp_data = pd.read_csv(exp_dir)
cluster_data = pd.read_excel(cluster_dir)
```

```python
j=0
plt.rcParams['figure.figsize'] = [15, 10]
for i in range(0, 4):
    plt.subplot(221+j)
    plt.plot(np.arange(8), exp_data.iloc[i][1:])
    plt.scatter(np.arange(8), exp_data.iloc[i][1:], c='r')
    xt = list(map(str, cluster_data.iloc[i][1:]))
    plt.xticks([])
    j+=1

plt.savefig('cluster_result/cluster_rel.png')
plt.show()
```
